```python
from scipy.spatial import cKDTree
```

```python
# Extract spatial coordinates
swe_coords = merged_swe[['Latitude', 'Longitude']].values
meteo_coords = merged_md[['lat', 'lon']].drop_duplicates().values

# Build KD-Tree using meteorological grid points
tree = cKDTree(meteo_coords)

# Find the nearest meteorological grid point for each SWE station
distances, indices = tree.query(swe_coords, k=1)

# Map each SWE station to its closest meteorological grid point
merged_swe['nearest_meteo_lat'] = meteo_coords[indices, 0]
merged_swe['nearest_meteo_lon'] = meteo_coords[indices, 1]
```

```python
merged_swe.rename(columns={'Date': 'date'}, inplace=True)
```

```python
# Rename for clarity before merging
merged_md = merged_md.rename(columns={'lat': 'nearest_meteo_lat', 'lon': 'nearest_meteo_lon'})

# Merge meteorological data with SWE data using the nearest matched lat/lon
final_data = pd.merge(
    merged_swe, merged_md,
    on=['date', 'nearest_meteo_lat', 'nearest_meteo_lon'],
    how='left'
)
```

```python
final_data
```

```python
final_data = pd.merge(final_data, merged_swe, on=['Latitude', 'Longitude'], how='left')
```

```python
import numpy as np

# Convert numerical data to float32 to save memory
merged_swe[['Latitude', 'Longitude']] = merged_swe[['Latitude', 'Longitude']].astype(np.float32)
final_data[['Latitude', 'Longitude']] = final_data[['Latitude', 'Longitude']].astype(np.float32)
```
